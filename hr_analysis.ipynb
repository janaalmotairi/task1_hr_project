```python
import pandas as pd
from sqlalchemy import create_engine

# first read the file
df= pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')

# clean the data
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
drop_columns=['EmployeeCount','StandardHours','Over18']
df.drop(columns=drop_columns,inplace=True,errors='ignore')

# create the database
engine=create_engine('sqlite:///hr_data.db')

# move the data from pandas to sql
df.to_sql('employees',con=engine,if_exists='replace',index=False)


pd.set_option('display.max_columns',None)
df.head()
```

```python
## Question 1: How many total employees are there?
q_total="SELECT COUNT(*) AS Total_Employees FROM employees"
total_df=pd.read_sql(q_total,con=engine)

print("the total employee count is: ")
total_df
```

```python
## Question 2: What is the employee count for each department?
q_total_by_department="""SELECT Department, COUNT (*) AS Employee_count 
FROM employees GROUP BY Department"""
total_by_department_df=pd.read_sql(q_total_by_department,con=engine)


print("the total number of employees by each department:")
total_by_department_df
```

```python
## Question 3: What is the average monthly income by job role?
q_income=""" SELECT JobRole, AVG(MonthlyIncome) AS Average_Income FROM employees 
GROUP BY JobRole"""
avg_income_df=pd.read_sql(q_income,con=engine)

print("the average income by job role:")
avg_income_df
```

```python
##  Question 4: Who are the top 5 employees by performance rate?
q_top_performance="""SELECT EmployeeNumber, Department, PerformanceRating FROM employees
ORDER BY PerformanceRating DESC LIMIT 5"""
top_performance_df=pd.read_sql(q_top_performance,con=engine)

print("the top 5 employees by performance rating: ")
top_performance_df
```

```python
## Question 5: Which department has the highest average performance rating?
q_top_department_performance=""" SELECT Department, AVG(PerformanceRating) AS Average_Performance_Rating_By_Department 
FROM employees
GROUP BY Department
ORDER BY PerformanceRating DESC
"""
top_avg_performance_df=pd.read_sql(q_top_department_performance,con=engine)

print("the highest average performance rating by department: ")
top_avg_performance_df
```

```python
## Question 6: What is the average job satisfaction rate for each department?
q_job_Satisfaction=""" SELECT Department, AVG(JobSatisfaction) FROM employees
GROUP BY Department
ORDER BY JobSatisfaction DESC
"""
job_Satisfaction_df=pd.read_sql(q_job_Satisfaction,con=engine)

print("the average job satisfaction rate by department: ")
job_Satisfaction_df
```

```python
## Question 7: Does working overtime affect the emplyees performance rate?
q_overtime=""" SELECT OverTime, AVG(PerformanceRating) AS Average_Performance
FROM employees
GROUP BY OverTime 
ORDER BY Average_Performance DESC"""
overtime_df=pd.read_sql(q_overtime,con=engine)

print("the average performance rating by overtime: ")
overtime_df
```

```python
## Question 8: What is the average Monthly Income for each gender?
q_avg_income_by_gender=""" SELECT Gender, AVG(MonthlyIncome) AS Average_Monthly_Income
FROM employees
GROUP BY Gender"""
avg_income_by_gender_df=pd.read_sql(q_avg_income_by_gender,con=engine)

print("average income for each gender: ")
avg_income_by_gender_df
```
